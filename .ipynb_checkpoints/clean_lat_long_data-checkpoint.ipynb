```python
import pandas as pd
import json
```

```python
st_data = pd.read_csv('station_lat_long.csv')

print(st_data.shape)

st_data.head()
```

```python
st_data_clean = st_data.rename(columns={'Longitude':'longitude',
                                        'Latitude':'latitude',
                                        'Station': 'NAME',
                                        'Region': 'AREA'})
st_data_clean.head()
```

```python
def df_to_geojson(df, properties, lat='latitude', lon='longitude'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson

st_geojson = df_to_geojson(st_data_clean,['NAME', 'AREA'])
```

```python
st_geojson
```

```python
with open('st_lat_long.geojson', 'w') as outfile:
    json.dump(st_geojson, outfile, ensure_ascii=False, indent=2)
```
